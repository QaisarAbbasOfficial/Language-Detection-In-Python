### Scraping Text Procedure

en - English,
ru - Russian,
ar - Arabic

```python
# Creating List For Each Language
eng = []
rus = []
ara = []

# Creating A Nested List
data_list = [eng,rus,ara]

# Language Notation For Wikipedia Search
language = ["en","ru","ar"]
```

```python
# Text Scraping Function
def text_scraping(topic):
    for i in range(len(language)):
        url = f"https://{language[i]}.wikipedia.org/wiki/{topic}"
        print(f"Fetching Data from Wikipedia In ({language[i]}): ", url)
        soup = BeautifulSoup(requests.get(url).text, "lxml")
        for paragraph in soup.find_all("p"):
            data_list[i].append(paragraph.text.strip())
```

```python
# Scraping The Details Of Wikipedia Search For "Natural Language Processing"
text_scraping("Natural Language Processing")
```
